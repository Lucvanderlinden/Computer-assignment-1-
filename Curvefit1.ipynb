```python
dts2= np.logspace(-6, -2, 50)

E_difference=[]
from scipy.optimize import curve_fit

for timestep in dts2:
    positions, momenta, time = velocity_verlet_integrator(
        initial_position=1.3,
        force_function=normalized_lj_force,
        dt=timestep
    )
    E, K, P=All_energies(momenta, positions)
    fluct_E=E.max()-E.min()
    E_difference.append(fluct_E)

def test_function(x, a,b):
    return a*x**b
dts2_array=np.array(dts2)
E_difference_array=np.array(E_difference)
values, cov = curve_fit(test_function, dts2_array, E_difference_array)

a_opt, b_opt = values
print(f"Optimale parameters: a={a_opt}, b={b_opt}")

plt.figure()
plt.grid()
plt.plot(dts2_array, E_difference_array, marker='.', linestyle="", label='Data', color='orangered')
plt.plot(dts2_array, test_function(dts2_array, *values), color='deepskyblue', label=fr'Fit ${a_opt:.0f}\, dt^{{{b_opt:.1f}}}$')
plt.xlabel('dt')
plt.ylabel('Energy fluctuation')
plt.loglog()
plt.legend()
plt.savefig(f'fluctuation vs dt.png', dpi=500)
plt.show()
```
